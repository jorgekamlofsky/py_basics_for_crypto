# Cifrado de Flujo con Vernam

## Introduccion

El cifrado de flujo pretende enviar información de un flujo de bits encriptados simétricamente correspondientes a la información que deseamos enviar en forma segura. <br>
El archivo creado se descompone en bits para que cada uno de ellos pueda ser cifrado "on the fly" y transmitido sin demoras. En la otra punta, el receptor recompone el archivo y lo desencripta. <br>
En esta Notebook se presenta un archivo, se lo descompone, se lo encripta, y al archivo encriptado se l desencrpta y finalmente se recompone el archivo.

## El cifrador Vernam

Vernam es un cifrador simético de flujo muy rápido y eficiente: sencillo e implementable por hardware y software. Es considerado "incondicionalmente seguro si la clave compartida se genera por métodos aleatorios y su longitud es igual a la del mensaje".<br>
A grandes razgos, consiste en el cifrado bit a bit de un flujo de información (que puede transmitirse en tiempo real sin demora) realizando una operación XOR ente el bit del texto claro con el respectivo bit de la clave (o flujo de clave).
